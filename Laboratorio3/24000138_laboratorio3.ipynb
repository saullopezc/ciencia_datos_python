1. ¿Cuántos Municipios tienen más de 100000 habitantes?.

```python
import pandas as pd
import matplotlib.pyplot as plt

archivo_csv = "./Lab3/NucleosPoblacion.csv"
datos = pd.read_csv(archivo_csv)

#print("Primeras filas del dataset:")
#print(datos.head())

#print("\nColumnas del dataset:")
#print(datos.columns)

municipios_mas_100k = datos[datos['Poblacion'] > 100000]


print(f"Son {len(municipios_mas_100k)} los municipios con más de 100,000 habitantes:")
for index, municipio in municipios_mas_100k.iterrows():
    nombre_municipio = municipio['Municipio']
    poblacion = municipio['Poblacion']
    print(f"{nombre_municipio}: {int(poblacion)} habitantes")
```

5. Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
municipios_granada = datos[datos['Provincia'] == 'Granada']

media_poblacion = municipios_granada['Poblacion'].mean()
mediana_poblacion = municipios_granada['Poblacion'].median()
desviacion_estandar_poblacion = municipios_granada['Poblacion'].std()
maximo_poblacion = municipios_granada['Poblacion'].max()
minimo_poblacion = municipios_granada['Poblacion'].min()

print("Estadísticas de la población de los municipios de la provincia de Granada:")
print(f"Media: {media_poblacion:.2f}")
print(f"Mediana: {mediana_poblacion:.2f}")
print(f"Desviación Estándar: {desviacion_estandar_poblacion:.2f}")
print(f"Valor Máximo: {maximo_poblacion}")
print(f"Valor Mínimo: {minimo_poblacion}")
```

8. Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python

# se uso pitagoras

municipios_no_madrid = datos[~((datos['Provincia'] == 'Madrid') & (datos['Municipio'] == 'Madrid'))]
#print(municipios_no_madrid)
madrid = datos[(datos['Provincia'] == 'Madrid') & (datos['Municipio'] == 'Madrid')]

coordenadas_municipios_madrid = madrid[['Y', 'X']]
print(coordenadas_municipios_madrid)

def distancia_pitagoras(lat1, lon1, lat2, lon2):
    x1 = lon1
    y1 = lat1
    x2 = lon2
    y2 = lat2
    distancia = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
    return distancia

# Calcular la distancia de cada municipio a Madrid
municipios_no_madrid['Distancia_a_Madrid'] = municipios_no_madrid.apply(
    lambda row: distancia_pitagoras(coordenadas_municipios_madrid['Y'], coordenadas_municipios_madrid['X'], row['Y'], row['X']),
    axis=1
)

municipio_mas_cercano = municipios_no_madrid.loc[municipios_no_madrid['Distancia_a_Madrid'].idxmin()]

municipio_mas_lejano = municipios_no_madrid.loc[municipios_no_madrid['Distancia_a_Madrid'].idxmax()]

print("Municipio más cercano a Madrid:")
print(f"Nombre: {municipio_mas_cercano['Municipio']}, Distancia: {municipio_mas_cercano['Distancia_a_Madrid']:.2f} km")

print("\nMunicipio más lejano a Madrid:")
print(f"Nombre: {municipio_mas_lejano['Municipio']}, Distancia: {municipio_mas_lejano['Distancia_a_Madrid']:.2f} km")

```
