```python
#extraemos faltantes de todas las columnas
for col in dataset.columns:
    porcentaje_na = dataset[col].isnull().mean()
    if(porcentaje_na > 0):
        if (dataset[col].dtype == 'object'):
            valor_a_imputar = dataset[col].mode()
        else:
            valor_a_imputar = dataset[col].mean()
        dataset[col].fillna(valor_a_imputar)
        print(f"{col}: {porcentaje_na}")
```

## Escala de variables

```python
len(dataset['children'].unique())
```

```python

def get_df_variable_scale_type(dataset):
    continuas = []
    discretas = []
    categoricas = []
    
    for col in dataset.columns:
        tipoDato = dataset[col].dtype
        if ((tipoDato == 'int64') or (tipoDato == 'float64')):
            if(len(dataset[col].unique()) > 30):
                continuas.append(col)
            else:
                discretas.append(col)
        else:
            categoricas.append(col)
    return categoricas , discretas , continuas
categoricas , discretas , continuas = get_df_variable_scale_type(dataset)
categoricas + discretas + continuas
```

```python
for col in continuas:
    estadisticos = dataset[col].describe()
    print()
    print(f"Columna: {col}")
    print(f"Media: {estadisticos['mean']} ")
    print(f"Minimo: {estadisticos['min']} ")
    print(f"Maximo: {estadisticos['max']} ")
    print(f"Mediana: {estadisticos['50%']} ")
    print(f"Desviacion: {estadisticos['std']} ")
    print(f"Varianza: {estadisticos['std']**2} ")
```
